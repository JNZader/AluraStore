### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/JNZader/AluraStore/refs/heads/main/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/JNZader/AluraStore/refs/heads/main/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/JNZader/AluraStore/refs/heads/main/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/JNZader/AluraStore/refs/heads/main/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

#1. Análisis de facturación



```python
#concatenar tiendas
tiendas = pd.concat([tienda, tienda2, tienda3, tienda4])

#calcular facturacion total
fact_total=tiendas['Precio'].sum()
print(f"La facturación total es de ${fact_total}")

#calcular facturacion por tienda
fact_x_tienda={
    'Tienda 1':tienda['Precio'].sum(),
    'Tienda 2':tienda2['Precio'].sum(),
    'Tienda 3':tienda3['Precio'].sum(),
    'Tienda 4':tienda4['Precio'].sum()
}
print("\nFacturación por tienda:")
for tienda, facturacion in fact_x_tienda.items():
    print(f"{tienda}: ${facturacion}")
```

# 2. Ventas por categoría

```python
from IPython.display import display

conteo_t1 = tienda.groupby("Categoría del Producto").size().rename("Tienda 1")
conteo_t2 = tienda2.groupby("Categoría del Producto").size().rename("Tienda 2")
conteo_t3 = tienda3.groupby("Categoría del Producto").size().rename("Tienda 3")
conteo_t4 = tienda4.groupby("Categoría del Producto").size().rename("Tienda 4")

conteo_total = pd.concat([conteo_t1, conteo_t2, conteo_t3, conteo_t4], axis=1).fillna(0).astype(int)

print("\nVentas por categoria en cada tienda:")
display(conteo_total)

categorias_mas_populares = conteo_total.idxmax()

categorias_mas_populares_df = categorias_mas_populares.reset_index()
categorias_mas_populares_df.columns = ['Tienda', 'Categoría Más Popular']

print("\nCategorías más populares por tienda:")
display(categorias_mas_populares_df)
```

# 3. Calificación promedio de la tienda


```python


# --- Tienda 1 ---
satisfaccion_tienda1 = tienda['Calificación'].mean()

# --- Tienda 2 ---
satisfaccion_tienda2 = tienda2['Calificación'].mean()

# --- Tienda 3 ---
satisfaccion_tienda3 = tienda3['Calificación'].mean()

# --- Tienda 4 ---
satisfaccion_tienda4 = tienda4['Calificación'].mean()

#crear un diccionario con los resultados
datos_satisfaccion = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio': [satisfaccion_tienda1, satisfaccion_tienda2, satisfaccion_tienda3, satisfaccion_tienda4]
}

#convertir el diccionario a un DataFrame de pandas
df_satisfaccion = pd.DataFrame(datos_satisfaccion)

#formatear la columna de calificacion
df_satisfaccion['Calificación Promedio'] = df_satisfaccion['Calificación Promedio'].map('{:.2f}'.format)


#mostrar el df
print("Resultados de Satisfacción del Cliente:")
display(df_satisfaccion)
```

# 4. Productos más y menos vendidos

```python
# --- Tienda 1 ---
conteo_productos_tienda1 = tienda['Producto'].value_counts()
mas_vendido_t1 = conteo_productos_tienda1.idxmax()
menos_vendido_t1 = conteo_productos_tienda1.idxmin()

print(f"--- Tienda 1 ---")
print(f"Producto más vendido: {mas_vendido_t1} ({conteo_productos_tienda1.max()} unidades)")
print(f"Producto menos vendido: {menos_vendido_t1} ({conteo_productos_tienda1.min()} unidades)")

# --- Tienda 2 ---
conteo_productos_tienda2 = tienda2['Producto'].value_counts()
mas_vendido_t2 = conteo_productos_tienda2.idxmax()
menos_vendido_t2 = conteo_productos_tienda2.idxmin()

print(f"--- Tienda 2 ---")
print(f"Producto más vendido: {mas_vendido_t2} ({conteo_productos_tienda2.max()} unidades)")
print(f"Producto menos vendido: {menos_vendido_t2} ({conteo_productos_tienda2.min()} unidades)")

# --- Tienda 3 ---
conteo_productos_tienda3 = tienda3['Producto'].value_counts()
mas_vendido_t3 = conteo_productos_tienda3.idxmax()
menos_vendido_t3 = conteo_productos_tienda3.idxmin()

print(f"--- Tienda 3 ---")
print(f"Producto más vendido: {mas_vendido_t3} ({conteo_productos_tienda3.max()} unidades)")
print(f"Producto menos vendido: {menos_vendido_t3} ({conteo_productos_tienda3.min()} unidades)")

# --- Tienda 4 ---
conteo_productos_tienda4 = tienda4['Producto'].value_counts()
mas_vendido_t4 = conteo_productos_tienda4.idxmax()
menos_vendido_t4 = conteo_productos_tienda4.idxmin()

print(f"--- Tienda 4 ---")
print(f"Producto más vendido: {mas_vendido_t4} ({conteo_productos_tienda4.max()} unidades)")
print(f"Producto menos vendido: {menos_vendido_t4} ({conteo_productos_tienda4.min()} unidades)")
```

# 5. Envío promedio por tienda

```python
# --- Tienda 1 ---
envio_promedio_t1 = tienda['Costo de envío'].mean()

# --- Tienda 2 ---
envio_promedio_t2 = tienda2['Costo de envío'].mean()

# --- Tienda 3 ---
envio_promedio_t3 = tienda3['Costo de envío'].mean()

# --- Tienda 4 ---
envio_promedio_t4 = tienda4['Costo de envío'].mean()

print(f"Costo de Envío Promedio Tienda 1: ${envio_promedio_t1:,.2f}")
print(f"Costo de Envío Promedio Tienda 2: ${envio_promedio_t2:,.2f}")
print(f"Costo de Envío Promedio Tienda 3: ${envio_promedio_t3:,.2f}")
print(f"Costo de Envío Promedio Tienda 4: ${envio_promedio_t4:,.2f}")
```
