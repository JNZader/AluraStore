### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/JNZader/AluraStore/refs/heads/main/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/JNZader/AluraStore/refs/heads/main/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/JNZader/AluraStore/refs/heads/main/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/JNZader/AluraStore/refs/heads/main/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

#1. Análisis de facturación



```python
#concatenar tiendas
tiendas = pd.concat([tienda, tienda2, tienda3, tienda4])

#calcular facturacion total
fact_total=tiendas['Precio'].sum()
print(f"La facturación total es de ${fact_total}")

#calcular facturacion por tienda
fact_x_tienda={
    'Tienda 1':tienda['Precio'].sum(),
    'Tienda 2':tienda2['Precio'].sum(),
    'Tienda 3':tienda3['Precio'].sum(),
    'Tienda 4':tienda4['Precio'].sum()
}
print("\nFacturación por tienda:")
for tienda, facturacion in fact_x_tienda.items():
    print(f"{tienda}: ${facturacion}")
```

# 2. Ventas por categoría

```python
import pandas as pd
from IPython.display import display

url = "https://raw.githubusercontent.com/JNZader/AluraStore/refs/heads/main/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/JNZader/AluraStore/refs/heads/main/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/JNZader/AluraStore/refs/heads/main/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/JNZader/AluraStore/refs/heads/main/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

conteo_t1 = tienda.groupby("Categoría del Producto").size().rename("Tienda 1")
conteo_t2 = tienda2.groupby("Categoría del Producto").size().rename("Tienda 2")
conteo_t3 = tienda3.groupby("Categoría del Producto").size().rename("Tienda 3")
conteo_t4 = tienda4.groupby("Categoría del Producto").size().rename("Tienda 4")

conteo_total = pd.concat([conteo_t1, conteo_t2, conteo_t3, conteo_t4], axis=1).fillna(0).astype(int)

print("\nVentas por categoria en cada tienda:")
display(conteo_total)

categorias_mas_populares = conteo_total.idxmax()

categorias_mas_populares_df = categorias_mas_populares.reset_index()
categorias_mas_populares_df.columns = ['Tienda', 'Categoría Más Popular']

print("\nCategorías más populares por tienda:")
display(categorias_mas_populares_df)
```

# 3. Calificación promedio de la tienda


```python
import pandas as pd
from IPython.display import display

url = "https://raw.githubusercontent.com/JNZader/AluraStore/refs/heads/main/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/JNZader/AluraStore/refs/heads/main/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/JNZader/AluraStore/refs/heads/main/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/JNZader/AluraStore/refs/heads/main/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# --- Tienda 1 ---
satisfaccion_tienda1 = tienda['Calificación'].mean()

# --- Tienda 2 ---
satisfaccion_tienda2 = tienda2['Calificación'].mean()

# --- Tienda 3 ---
satisfaccion_tienda3 = tienda3['Calificación'].mean()

# --- Tienda 4 ---
satisfaccion_tienda4 = tienda4['Calificación'].mean()

#crear un diccionario con los resultados
datos_satisfaccion = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio': [satisfaccion_tienda1, satisfaccion_tienda2, satisfaccion_tienda3, satisfaccion_tienda4]
}

#convertir el diccionario a un DataFrame de pandas
df_satisfaccion = pd.DataFrame(datos_satisfaccion)

#formatear la columna de calificacion
df_satisfaccion['Calificación Promedio'] = df_satisfaccion['Calificación Promedio'].map('{:.2f}'.format)


#mostrar el df
print("Resultados de Satisfacción del Cliente:")
display(df_satisfaccion)
```
